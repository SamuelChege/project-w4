```
#reading our csv file
df = pd.read_csv('clean_df.csv')
df.head(5)
```

```
#Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.
pop_hour =df[['Bluecar counter','City','hour']]
pop = pop_hour[pop_hour['City']=='Paris']
pop['hour'].value_counts().sort_values(ascending = False).head(1)
```

```
#What is the most popular hour for returning cars
df.columns.values.tolist()
```

```
#What station is the most popular?

pop_station = df[['Bluecar counter','City','hour', 'Kind', 'ID']]
station = pop_station[pop_station['City']=='Paris']
station['ID'].value_counts().sort_values(ascending=False).head(1)
```

```
#What postal code is the most popular for picking up Blue cars? Does the most popular station belong to that postal code?

postal_code = df[['Bluecar counter','City','hour', 'Postal code', 'Kind', 'ID']]
pop = postal_code[postal_code['ID'] == 'paris-portedemontrouge-8']
pop['Postal code'].value_counts().sort_values(ascending=False).head(1)
```
